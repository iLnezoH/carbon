```python
from pc import PC
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data = np.loadtxt('mock/mutualism.csv', delimiter=',')
data = np.loadtxt('mock/test.csv', delimiter=',')
data.shape
X = data[0]
Y = data[1]
```

```python
fig, axe = plt.subplots(figsize=(6, 6))
axe.set_xlim(0, 210)
axe.set_ylim(0, 130)
axe.plot(np.arange(200), X)
axe.plot(np.arange(200), Y)
```

```python
pxy = []
pyx = []

E = 2
tau = 1
for l in range(10, 200):
    res = PC(X[:l], Y[:l], E, tau)
    pxy.append(res[0][0])
    pyx.append(res[1][0])
```

```python
fig, axe = plt.subplots(figsize=(6,6))
xs = np.arange(190)
axe.plot(xs, pxy)
axe.plot(xs, pyx)
plt.show()
```
