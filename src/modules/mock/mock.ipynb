```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mock():
    L = 200
    X = np.empty(L)
    Y = np.empty(L)
    X[0] = 10
    Y[0] = 5
    K = 100

    def phi(t):
        p = np.random.random(1)[0]
        if p >= 0.5:
            return 0
        else:
            return 1
    for t in range(L - 1):
        p = phi(t)
        X[t + 1] = X[t] + 0.2 * X[t] * (1 - X[t] / K + p * 0.1 * Y[t] / K)
        Y[t + 1] = Y[t] + 0.2 * Y[t] * (1 - Y[t] / K + p * 0.2 * Y[t] / K)

    return X, Y
```

```python
X,Y= mock()
```

```python
data = np.vstack([X, Y])
data.shape
```

```python

np.savetxt('mutualism.csv', data, delimiter=',')

fig, axe = plt.subplots(figsize=(6, 6))
axe.set_xlim(-10, 210)
axe.set_ylim(50, 130)
axe.plot(np.arange(200), X)
axe.plot(np.arange(200), Y)
```
